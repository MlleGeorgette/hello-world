```python
""" In this tutorial, I call an API to retrieve a photo of Mars.
Some API requests, such as NASA, require authentication.
Source: https://realpython.com/python-api/ """
```

```python
# Import module
import requests
```

```python
# Creating parameters for the request
endpoint = "https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos"
api_key = "DEMO_KEY"
query_params = {"api_key": api_key, "earth_date": "2022-01-01"}
```

```python
# Make the request
response = requests.get(endpoint, params=query_params)
```

```python
# Check if it's successful - for instance, 401 means unauthorised; 404 means error
response
```

```python
# Look at response object
response.json()
```

```python
# Extract photos from it
photos = response.json()["photos"]
print(f"Found {len(photos)} photos")
```
